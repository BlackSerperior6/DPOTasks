```python
import math
import random
import matplotlib.pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection="3d")

# !
random.seed(2025)

CLR_0 = '#22ff33'
MARKER_1 = '^'
MARKER_2 = 'o'

K = 400
XLIM = 400
YLIM = 400
STEP = 4

def init():
    k=K
    Z = [i for i in range(k)]
    X = [math.cos(i/3)*(k-i) for i in range(k)]
    Y = [math.sin(i/3)*(k-i) for i in range(k)]
    ax.scatter(X,Y,Z, c=CLR_0, marker=MARKER_1)
    step = STEP
    clr = [(abs(random.random()-i/k),abs(random.random()-i/k),abs(random.random()-i/k)) for i in range(1,k,step)]
    Z = [i for i in range(1,k,step)]
    X = [math.cos(i/2+2)*(k-i+10) for i in range(1,k,step)]
    Y = [math.sin(i/2+2)*(k-i+10) for i in range(1,k,step)]
    ax.scatter(X,Y,Z, c=clr, marker=MARKER_2,s=40)
    plt.xlim(-XLIM,XLIM)
    plt.ylim(-YLIM,YLIM)
    return fig,

def animate(f):
    fig.clear()
    ax = fig.add_subplot(111, projection="3d")
    k=K
    Z = [i for i in range(k)]
    X = [math.cos(i/3+f/40)*(k) for i in range(k)]
    Y = [math.sin(i/3+f/40)*(k) for i in range(k)]
    ax.scatter(X,Y,Z, c=CLR_0, marker=MARKER_1)
    step = STEP
    clr = [(abs(random.random()-i/k),abs(random.random()-i/k),abs(random.random()-i/k)) for i in range(1,k,step)]
    Z = [i for i in range(1,k,step)]
    X = [math.cos(i/5+2+f/10)*(k-i+80) for i in range(1,k,step)]
    Y = [math.sin(i/5+2+f/10)*(k-i+80) for i in range(1,k,step)]
    ax.scatter(X,Y,Z, c=clr, marker=MARKER_2,s=40)
    plt.xlim(-XLIM,XLIM)
    plt.ylim(-YLIM,YLIM)
    return fig,

ani=animation.FuncAnimation(fig, animate, init_func=init,
                               frames=180, interval=50, blit=False)

gif_writer = animation.PillowWriter(fps=30)

ani.save('filename.gif',writer=gif_writer)
```
